```python
from transformers import BertForSequenceClassification, BertTokenizer
import pandas as pd
import torch
import torch.nn.functional as F
from tqdm import tqdm
```

```python
model = BertForSequenceClassification.from_pretrained("/Users/macbook/Documents/PhD_Documents/crudeBERT/model")
tokenizer = BertTokenizer.from_pretrained("/Users/macbook/Documents/PhD_Documents/crudeBERT/tokenizer")
```

```python
df_news = pd.read_csv('news_data.csv')
```

```python
df_news = pd.read_csv('/Users/macbook/Documents/PhD_Documents/Second_paper/new_data_2021/headlines/total_news_data/news_total_file.csv')
df_news
```

```python
# Convert the column to a list of sentences
sentences = df_news["Text"].tolist()

# Initialize an empty list to store sentiment scores
all_sentiment_scores = []

# Use tqdm to show the progress bar
for sentence in tqdm(sentences, desc="Processing Sentences"):
    # Tokenize each sentence individually
    tokenized = tokenizer([sentence], padding=True, truncation=True, return_tensors="pt")
    
    # Pass the tokenized input to the model
    outputs = model(**tokenized)
    
    # Get the sentiment scores using softmax
    predictions = F.softmax(outputs.logits, dim=1)
    
    # Extract the raw scores for each sentiment class
    sentiment_scores = {f"sentiment_{i}": score.item() for i, score in enumerate(predictions[0])}
    
    # Append the scores to the list
    all_sentiment_scores.append(sentiment_scores)

# Add sentiment scores to your dataframe
df_news["sentiment_scores"] = all_sentiment_scores
df_news
```

```python
# Function to map sentiment scores to a score between -1 and 1
def map_sentiment_score(sentiment_scores):
    weighted_sum = sentiment_scores[0] * 1 + sentiment_scores[1] * -1 + sentiment_scores[2] * 0
    total_weight = sum(sentiment_scores)
    normalized_score = weighted_sum / total_weight
    return normalized_score
```

```python
tqdm.pandas(desc="Processing Sentiment Scores")

# Apply the map_sentiment_score function to the sentiment_scores column in the dataframe with progress bar
df_news['normalized_sentiment_score'] = df_news['sentiment_scores'].progress_apply(
    lambda x: map_sentiment_score([x['sentiment_0'], x['sentiment_1'], x['sentiment_2']])
)

# Show the updated dataframe
df_news
```

```python
df_news = df_news.drop(['sentiment_scores', 'Text'], axis = 1)
df_news
```

```python
crudebert = df_news.groupby('Date')['normalized_sentiment_score'].mean().reset_index()

# Rename columns
crudebert.columns = ['Date', 'crudebert']
#crudebert = crudebert.rename(columns={'date': 'date_column'})
crudebert
```
